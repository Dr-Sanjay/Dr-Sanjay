```python
import matplotlib.pyplot as plt

# Sample data
repo_names = ['Repo 1', 'Repo 2', 'Repo 3', 'Repo 4']
commit_counts = [150, 300, 200, 100]
star_counts = [10, 50, 30, 20]
fork_counts = [5, 10, 8, 6]

# Bar chart
plt.figure(figsize=(8, 6))
plt.bar(repo_names, commit_counts, color='blue')
plt.xlabel('Repositories')
plt.ylabel('Commit Counts')
plt.title('Commit Counts by Repository')
plt.xticks(rotation=45)
plt.show()

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(star_counts, fork_counts, color='red')
plt.xlabel('Star Counts')
plt.ylabel('Fork Counts')
plt.title('Star Counts vs Fork Counts')
plt.show()
```
